```python
# basic set operations
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
set_a.add(7)
print(set_a)  # Output: {1, 2, 3, 4, 7}
set_a.remove(2)
print(set_a)  # Output: {1, 3, 4, 7}
```

```python
# removing an element 
set_a.remove(3)
print(set_a)  # Output: {1, 4, 7}
```

```python
# dont want error when removing an element that may not exist
set_a.discard(10)  # No error, even though 10 is not in set_a
print(set_a)  # Output: {1, 4, 7}   
```

```python
# pop method
popped_element = set_a.pop()
print(popped_element)  # Output: 1 (or 4 or 7, since sets are unordered)
print(set_a)  # Output: {4, 7} (or {1, 7} or {1, 4}, depending on which element was popped
```

```python
# clear
set_a.clear()
print(set_a)  # Output: set()
```

```python
# set membership
print(3 in set_b)  # Output: True
print(10 in set_b)  # Output: False
```

```python
# mathematical set operations
set_c = {1, 2, 3}
set_d = {3, 4, 5}
union_set = set_c.union(set_d)
print(union_set)  # Output: {1, 2, 3, 4, 5}
```

```python
# intersection
intersection_set = set_c.intersection(set_d)
print(intersection_set)  # Output: {3}
```

```python
# intersection update
set_c.intersection_update(set_d)
print(set_c)  # Output: {3}
```

```python
# difference
difference_set = set_d.difference(set_c)
print(difference_set)  # Output: {4, 5} 
```

```python
# difference update
set_d.difference_update(set_c)
print(set_d)  # Output: {4, 5}
```

```python
# syymetric difference
set_e = {1, 2, 3}
set_f = {3, 4, 5}
sym_diff_set = set_e.symmetric_difference(set_f)
print(sym_diff_set)  # Output: {1, 2, 4, 5}
```

```python
# subset and superset
set_g = {1, 2}
set_h = {1, 2, 3, 4}
print(set_g.issubset(set_h))  # Output: True
print(set_h.issuperset(set_g))  # Output: True
```
