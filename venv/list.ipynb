```python
### accessing list elements
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
print(fruits[0])  # Output: apple
print(fruits[2:4])  # Output: cherry, date
```

```python
# modifying list elements
fruits[1] = "blueberry"
print(fruits)  # Output: ['apple', 'blueberry', 'cherry',
```

```python
# list methods
fruits.append("fig")
print(fruits)  # Output: ['apple', 'blueberry', 'cherry', 'date', 'elderberry', 'fig']
```

```python
fruits.insert(2, "coconut")
print(fruits)  # Output: ['apple', 'blueberry', 'coconut', 'cherry', 'date', 'elderberry', 'fig']
```

```python
# remove and return the last item
last_fruit = fruits.pop()
print(last_fruit)  # Output: fig
```

```python
index = fruits.index("cherry")  # Output: 3
print(index)
```

```python
fruits.insert(2, "coconut")
print(fruits)  # Output: ['apple', 'blueberry', 'coconut', 'cherry', 'date', 'elderberry', 'fig']
```

```python
count = fruits.count("coconut")  # Output: 2
print(count)
```

```python
fruits.sort()
print(fruits)  # Output: ['apple', 'blueberry', 'cherry', 'coconut', 'date', 'elderberry']
```

```python
# for reversing the list
fruits.reverse()
print(fruits)  # Output: ['elderberry', 'date', 'coconut', 'cherry', 'blueberry', 'apple']
```
