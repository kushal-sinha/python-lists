```python
mixed_tuple = (1, "Hello", 3.14, [1, 2, 3], (4, 5))
print(mixed_tuple)  # Output: (1, 'Hello', 3.14, [1, 2, 3], (4, 5))
```

```python
# accessing elements in a tuple
my_tuple = (1, "Hello", 3.14, [1, 2, 3], (4, 5))
print(my_tuple[0])  # Output: 1
print(my_tuple[1])  # Output: Hello
print(my_tuple[2])  # Output: 3.14
print(my_tuple[3])  # Output: [1, 2, 3]
print(my_tuple[4])  # Output: (4, 5)
```

```python
# tuple operations
tuple1 = (1, 2, 3)
tuple2 = ("single", "double", "triple")
# concatenation
tuple3 = tuple1 + tuple2
print(tuple3)  # Output: (1, 2, 3, 4, 5, 6)
# repetition
```

```python
# repetition
tuple4 = tuple1 * 2
print(tuple4)  # Output: (1, 2, 3, 1, 2, 3)
```

```python
# immutability of tuples
my_tuple = (1, 2, 3)
# my_tuple[0] = 10  # This will raise a TypeError: 'tuple' object does not support item assignment
```

```python
## tuple methods
my_tuple = (1, 2, 3, 4, 5)
print(my_tuple.count(2))  # Output: 1
print(my_tuple.index(3))  # Output: 2
```

```python
# packing and unpacking
packed_tuple = 1, "Hello", 3.14
print(packed_tuple)  # Output: (1, 'Hello', 3.14)
```

```python
# unpacking
a, b, c = packed_tuple
print(a)  # Output: 1
print(b)  # Output: Hello
print(c)  # Output: 3.14
```

```python
 # unpacking with *
numbers = (1, 2, 3, 4, 5)
a, *b, c = numbers
print(a)  # Output: 1
print(b)  # Output: [2, 3, 4]
print(c)  # Output: 5
```

```python
# examples of nested tuples
nested_tuple = (1, (2, 3), (4, 5, 6))
print(nested_tuple)  # Output: (1, (2, 3), (4, 5, 6))
print(nested_tuple[1])  # Output: (2, 3)
print(nested_tuple[1][0])  # Output: 2
print(nested_tuple[2][1])  # Output: 5
```

```python
# iterating through a tuple
my_tuple = (1, 2, 3, 4, 5)
for item in my_tuple:
    print(item)  # Output: 1 2 3 4 5 (each on a new line)
```
